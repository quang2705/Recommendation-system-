- Next I wrote a function to query approximately 7000 movies in the database. I have to query 40 movies at the time and stop
for 10 second because MovieDB does not allow user to query too much movie at the a time.
- Even with 7000 movies, this process should take at least 30 mins
- For bigger dataset, if you are not running it on a server, I suggest break it down to several csv file each day and then combie it together.

```python
def getting_small_movie(url, D):
    movie_list= []
    for index in range (0, 200):
        for movieID in range(0,40):
            api = requests.get(url.format(movieID + index*40),params = D)
            if(api.status_code == 200):
                movie = api.json()
                movie_list.append(movie)
        time.sleep(10)

    movie_data = pd.DataFrame(movie_list)
    average_rate = movie_data.vote_average.mean()
    movie_data['average_rate'] = average_rate
    return movie_data

movie_data = getting_small_movie(url, D)
movie_data.to_csv("data/7000movie_data.csv")
```

## Cleanning, correcting datatype 
- Since I am also using this dataset for a Django app, which require you to have a correct datatype for each column. Moreover, it is gennerally a good practice to convert each column into a same datatype

```python
movie = pd.read_csv("data/md.csv")
```

```python
md = movie[['id','title','overview', 'tagline', 'vote_average', 'vote_count', 'weight_rate','actors','director','details']]
md = md.dropna()

md['title'] = md['title'].astype(str)
md['id'] = md['id'].astype(int)
md['overview'] = md['overview'].astype(str)
md['tagline'] = md['tagline'].astype(str)
md['vote_average'] = md['vote_average'].astype(float)
md['vote_count'] = md['vote_count'].astype(int)
md['weight_rate'] = md['weight_rate'].astype(float)
md['actors'] = md['actors'].astype(str)
md['director'] = md['director'].astype(str)
md['details'] = md['details'].astype(str)
md
```

```python
md.to_csv('data/md.csv')
```

```python
link = pd.read_csv('data/links_small.csv')
link.dropna()
link['movieId'] = link['movieId'].astype(int)
link['imdbId'] = link['imdbId'].astype(int)
link['tmdbId'] = link['tmdbId'].astype(float)

for index, li in link[['tmdbId']].iterrows():
    if li[0] != li[0]:
        link = link.drop(index)
        
link['tmdbId'] = link['tmdbId'].astype(int)
link= link[['movieId','imdbId', 'tmdbId']]
link.to_csv('data/links_small.csv')
link.head(5)
```
