```python
import pandas as pd
import csv 
```

```python
movie = pd.read_csv('tmdb_movie.csv')
movie
```

```python
md = movie[['id','title','overview', 'tagline', 'vote_average', 'vote_count', 'weight_rate','actors','director','details']]
md = md.dropna()
# md = md.dropna(subset = ['title', 'vote_count', 'vote_average', 'weight_rate'])
md['title'] = md['title'].astype(str)
md['id'] = md['id'].astype(int)
md['overview'] = md['overview'].astype(str)
md['tagline'] = md['tagline'].astype(str)
md['vote_average'] = md['vote_average'].astype(float)
md['vote_count'] = md['vote_count'].astype(int)
md['weight_rate'] = md['weight_rate'].astype(float)
md['actors'] = md['actors'].astype(str)
md['director'] = md['director'].astype(str)
md['details'] = md['details'].astype(str)

md
```

```python
link = pd.read_csv('links_small.csv')

link.dropna()

link['movieId'] = link['movieId'].astype(int)
link['imdbId'] = link['imdbId'].astype(int)
link['tmdbId'] = link['tmdbId'].astype(float)

for index, li in link[['tmdbId']].iterrows():
    if li[0] != li[0]:
        link = link.drop(index)
        
link['tmdbId'] = link['tmdbId'].astype(int)
link= link[['movieId','imdbId', 'tmdbId']]
link.to_csv('links_small.csv')
link
```
